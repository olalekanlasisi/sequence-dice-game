```python
from random import randint, choice
import statistics
from itertools import product
import pandas as pd
import numpy as np
import pygame, sys
from pygame.locals import *
import pygame.surfarray as surfarray
square_size = 100
column_size = 6
row_size = 6
WHITE = (255, 255, 255)
LIGHT_BLUE = (56, 182, 255)
GREEN = (0,255,0)
YELLOW = (255,255,0)
square_size = 100
RADIUS = int(square_size/2 -10)
width = column_size * square_size
height = (row_size+1) * square_size
size = (width, height)
pygame.font.init()
 
# Total window
screen = pygame.display.set_mode((600, 900))
screen.fill(LIGHT_BLUE)
val = 0
# Load test fonts for future use
font1 = pygame.font.SysFont("comicsans", 40)
font2 = pygame.font.SysFont("comicsans", 20)
font3 = pygame.font.SysFont("Calibri", 25)
font4 = pygame.font.SysFont("comicsans", 20)

dice1 = pygame.image.load("dice_1.png")
dice2 = pygame.image.load("dice_2.png")
dice3 = pygame.image.load("dice_3.png")
dice4 = pygame.image.load("dice_4.png")
dice5 = pygame.image.load("dice_5.png")
dice6 = pygame.image.load("dice_6.png")
dice_face = [None, dice1, dice2, dice3, dice4, dice5, dice6]


def displayBoard(state):
    print("[",state[0][0]," |",state[0][1], " |",state[0][2], " |",state[0][3]," |",state[0][4]," |",state[0][5]," ]","\n\n")
    print("[",state[1][0]," |",state[1][1], " |",state[1][2], " |",state[1][3]," |",state[1][4]," |",state[1][5]," ]","\n\n")
    print("[",state[2][0]," |",state[2][1], " |",state[2][2], "|",state[2][3],"|",state[2][4]," |",state[2][5]," ]","\n\n")
    print("[",state[3][0]," |",state[3][1], " |",state[3][2], "|",state[3][3],"|",state[3][4]," |",state[3][5]," ]","\n\n")
    print("[",state[4][0]," |",state[4][1], " |",state[4][2], " |",state[4][3]," |",state[4][4]," |",state[4][5]," ]","\n\n")
    print("[",state[5][0]," |",state[5][1], " |",state[5][2], " |",state[5][3]," |",state[5][4]," |",state[5][5]," ]","\n\n")


def defensive_strategy_two_players(state, diceRoll, player):
    placed = False
    row = False
    column = False
    diagonal = False
    diagonal_inverse = False
    chip = ''
    if player == 0:
        player = 'X'
        chip = 'O'
    else:
        player ='O'
        chip = 'X'
    max_count_row, max_count_column, max_count_diagonal, max_count_diagonal_inverse = 0, 0, 0,0
    row,column,diagonal,diagonal_inverse = 0,0,0,0
    for length in range(-6, len(state)):
        len_row = tuple(state[length,:]).count(chip)
        len_column = tuple(state[:,length]).count(chip)
        len_diagonal = tuple(state.diagonal(length)).count(chip)
        len_diagonal_inverse = tuple(np.flipud(state).diagonal(length)).count(chip)

        if len_row > max_count_row:
            max_count_row = len_row
        if len_column > max_count_column:
            max_count_column = len_column
        if len_diagonal > max_count_diagonal:
            max_count_diagonal = len_diagonal
        if len_diagonal_inverse > max_count_diagonal_inverse:
            max_count_diagonal_inverse = len_diagonal_inverse

        if max_count_row >= max_count_column and max_count_row >= max_count_diagonal and max_count_row >= max_count_diagonal_inverse and diceRoll in state[length,:]:
            row = True
        elif tuple(state[length,:]).count(chip)>=2  and  diceRoll in state[length,:]:
            row = True
        elif max_count_column >= max_count_row and max_count_column >= max_count_diagonal and max_count_column >= max_count_diagonal_inverse and diceRoll in state[:,length]:
            column = True
        elif tuple(state[length,:]).count(chip)>=2  and  diceRoll in state[length,:]:
            column = True
        elif max_count_diagonal >= max_count_row and max_count_diagonal >= max_count_column and max_count_diagonal >= max_count_diagonal_inverse and diceRoll in state.diagonal(length):
            diagonal = True
        elif tuple(state.diagonal(length)).count(chip)>=2 and diceRoll in state.diagonal(length):
            diagonal = True
        elif max_count_diagonal_inverse >= max_count_row and max_count_diagonal_inverse >= max_count_column and max_count_diagonal_inverse >= max_count_diagonal and diceRoll in np.flipud(state).diagonal(length):
            diagonal_inverse = True
        elif tuple(np.flipud(state).diagonal(length)).count(chip)>=2 and diceRoll in np.flipud(state).diagonal(length):
            diagonal_inverse = True
        if row == True:
            dice = np.argwhere(diceRoll == state[length,:])
            index = dice[0][0]
            state[length,:][index] = player
            placed = True
            break
        elif column == True:
            dice = np.argwhere(diceRoll == state[:,length])
            index = dice[0][0]
            state[:,length][index] = player
            placed = True
            break
            
        elif diagonal == True:
            dice = np.argwhere(diceRoll == state.diagonal(length))
            index = dice[0][0]
            diag = state.diagonal(length).copy()
            diag[index] = player
            if len(diag) < 6 and length > 0:
                    np.fill_diagonal(state[:-1, length:], diag)
                    placed = True
                    break

            elif len(diag) < 6 and length < 0:
                np.fill_diagonal(state[abs(length):, :-1], diag)
                placed = True
                break

            else:
                np.fill_diagonal(state, diag)
                placed = True
                break
                
        elif diagonal_inverse == True:
            dice = np.argwhere(diceRoll == np.flipud(state).diagonal(length))
            index = dice[0][0]
            diag = np.flipud(state).diagonal(length).copy()
            diag[index] = player
            if len(diag) < 6 and length > 0:
                np.fill_diagonal(np.flipud(state)[:-1, length:], diag)
                placed = True
                break

            elif len(diag) < 6 and length < 0:
                np.fill_diagonal(np.flipud(state)[abs(length):, :-1], diag)
                placed = True
                break

            else:
                np.fill_diagonal(np.flipud(state), diag)
                placed = True
                break
       
    if placed == False:
        dice = np.argwhere(diceRoll == state)
        index = tuple(choice(dice))
        state[index] = player
    return state
    
    
def offensive_strategy_two_players(state, diceRoll, player):
    placed = False
    row = False
    column = False
    diagonal = False
    diagonal_inverse = False
    if player == 0: player = 'X'
    elif player == 1: player = 'O'
    max_count_row, max_count_column, max_count_diagonal, max_count_diagonal_inverse = 0, 0, 0,0
    row,column,diagonal,diagonal_inverse = 0,0,0,0
    for length in range(-6, len(state)):
        len_row = tuple(state[length,:]).count(player)
        len_column = tuple(state[:,length]).count(player)
        len_diagonal = tuple(state.diagonal(length)).count(player)
        len_diagonal_inverse = tuple(np.flipud(state).diagonal(length)).count(player)

        if len_row > max_count_row:
            max_count_row = len_row
        if len_column > max_count_column:
            max_count_column = len_column
        if len_diagonal > max_count_diagonal:
            max_count_diagonal = len_diagonal
        if len_diagonal_inverse > max_count_diagonal_inverse:
            max_count_diagonal_inverse = len_diagonal_inverse

        if max_count_row >= max_count_column and max_count_row >= max_count_diagonal and max_count_row >= max_count_diagonal_inverse and diceRoll in state[length,:]:
            row = True
        elif tuple(state[length,:]).count(player)>=2  and  diceRoll in state[length,:]:
            row = True
        elif max_count_column >= max_count_row and max_count_column >= max_count_diagonal and max_count_column >= max_count_diagonal_inverse and diceRoll in state[:,length]:
            column = True
        elif tuple(state[length,:]).count(player)>=2  and  diceRoll in state[length,:]:
            column = True
        elif max_count_diagonal >= max_count_row and max_count_diagonal >= max_count_column and max_count_diagonal >= max_count_diagonal_inverse and diceRoll in state.diagonal(length):
            diagonal = True
        elif tuple(state.diagonal(length)).count(player)>=2 and diceRoll in state.diagonal(length):
            diagonal = True
        elif max_count_diagonal_inverse >= max_count_row and max_count_diagonal_inverse >= max_count_column and max_count_diagonal_inverse >= max_count_diagonal and diceRoll in np.flipud(state).diagonal(length):
            diagonal_inverse = True
        elif tuple(np.flipud(state).diagonal(length)).count(player)>=2 and diceRoll in np.flipud(state).diagonal(length):
            diagonal_inverse = True
        if row == True:
            dice = np.argwhere(diceRoll == state[length,:])
            index = dice[0][0]
            state[length,:][index] = player
            placed = True
            break
        elif column == True:
            dice = np.argwhere(diceRoll == state[:,length])
            index = dice[0][0]
            state[:,length][index] = player
            placed = True
            break
            
        elif diagonal == True:
            dice = np.argwhere(diceRoll == state.diagonal(length))
            index = dice[0][0]
            diag = state.diagonal(length).copy()
            diag[index] = player
            if len(diag) < 6 and length > 0:
                    np.fill_diagonal(state[:-1, length:], diag)
                    placed = True
                    break

            elif len(diag) < 6 and length < 0:
                np.fill_diagonal(state[abs(length):, :-1], diag)
                placed = True
                break

            else:
                np.fill_diagonal(state, diag)
                placed = True
                break
                
        elif diagonal_inverse == True:
            dice = np.argwhere(diceRoll == np.flipud(state).diagonal(length))
            index = dice[0][0]
            diag = np.flipud(state).diagonal(length).copy()
            diag[index] = player
            if len(diag) < 6 and length > 0:
                np.fill_diagonal(np.flipud(state)[:-1, length:], diag)
                placed = True
                break

            elif len(diag) < 6 and length < 0:
                np.fill_diagonal(np.flipud(state)[abs(length):, :-1], diag)
                placed = True
                break

            else:
                np.fill_diagonal(np.flipud(state), diag)
                placed = True
                break
       
    if placed == False:
        dice = np.argwhere(diceRoll == state)
        index = tuple(choice(dice))
        state[index] = player
    return state
    

def cells_three_decision(state, player, diceRoll, skipped_turn, strategy):
    dR3 = np.argwhere(3 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 3 and state[0][1] == player and state[1][5] == player and state[4][0] == player and state[5][4] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
    
    if diceRoll == 3 and state[0][1] == chip and state[1][5] == chip and state[4][0] == chip and state[5][4] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n")
    
    elif len(dR3) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")

    elif state[0][1] == 3 or state[1][5] == 3 or state[4][0] == 3 or state[5][4] == 3:
        if len(dR3) == 1:
            text2 = font3.render("Place chip on last 3 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 3 left", "\n")
            dice3 = tuple(choice(dR3))
            state[dice3] = player
            displayBoard(state)
            print("\n\n")
            
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 3s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 3s using random strategy", "\n")
                dice3 = tuple(choice(dR3))
                state[dice3] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 3s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 3s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 3s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 3s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
                
    return state, skipped_turn
            
def cells_four_decision(state, player, diceRoll, skipped_turn, strategy):
    dR4 = np.argwhere(4 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 4 and state[0][2] == player and state[2][5] == player and state[3][0] == player and state[5][3] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")

    if diceRoll == 4 and state[0][2] == chip and state[2][5]  == chip and state[3][0] == chip and state[5][3] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n")
    
    elif len(dR4) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")
            
    elif state[0][2] == 4 or state[2][5] == 4 or state[3][0]  == 4 or state[5][3] == 4:
        if len(dR4) == 1:
            text2 = font3.render("Place chip on last 4 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 4 left", "\n")
            dice4 = tuple(choice(dR4))
            state[dice4] = player
            displayBoard(state)
            print("\n\n")

        else:
        
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 4s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 4s using random strategy", "\n")
                dice4 = tuple(choice(dR4))
                state[dice4] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 4s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 4s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 4s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 4s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn

def cells_five_decision(state, player, diceRoll, skipped_turn, strategy):
    dR5 = np.argwhere(5 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 5 and state[0][3] == player and state[2][0] == player and state[3][5] == player and state[5][2] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
    
    if diceRoll == 5 and state[0][3] == chip and state[2][0]  == chip and state[3][5] == chip and state[5][2] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n")
            
    elif len(dR5) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")
    
    elif state[0][3] == 5 or state[2][0] ==5 or state[3][5] == 5 or state[5][2] == 5:
        if len(dR5) == 1:
            text2 = font3.render("Place chip on last 5 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 5 left", "\n")
            dice5 = tuple(choice(dR5))
            state[dice5] = player
            displayBoard(state)
            print("\n\n")

        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 5s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 5s using random strategy", "\n")
                dice5 = tuple(choice(dR5))
                state[dice5] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 5s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 5s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 5s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 5s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn

def cells_six_decision(state, player, diceRoll, skipped_turn, strategy):
    dR6 = np.argwhere(6 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 6 and state[0][4] == player and state[1][0] == player and state[4][5] == player and state[5][1] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
    
    if diceRoll == 6 and state[0][4] == chip and state[1][0]  == chip and state[4][5] == chip and state[5][1] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n")
    
    elif len(dR6) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")
        
    elif state[0][4] == 6 or state[1][0] == 6 or state[4][5] == 6 or state[5][1] == 6:
        if len(dR6) == 1:
            text2 = font3.render("Place chip on last 6 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 6 left", "\n")
            dice6 = tuple(choice(dR6))
            state[dice6] = player
            displayBoard(state)
            print("\n\n")

        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 6s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 6s using random strategy", "\n")
                dice6 = tuple(choice(dR6))
                state[dice6] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 6s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 6s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 6s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 6s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn


def cells_seven_decision(state, player, diceRoll, skipped_turn, strategy):
    dR7 = np.argwhere(7 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 7 and state[1][1] == player and state[1][4] == player and state[4][1] == player and state[4][4] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
     
    if diceRoll == 7 and state[1][1] == chip and state[1][4]  == chip and state[4][1] == chip and state[4][4] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n") 
    
    elif len(dR7) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")
            
    elif state[1][1] == 7 or state[1][4] == 7 or state[4][1] == 7 or state[4][4] == 7:
        if len(dR7) == 1:
            text2 = font3.render("Place chip on last 7 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 7 left", "\n")
            dice7 = tuple(choice(dR7))
            state[dice7] = player
            displayBoard(state)
            print("\n\n")
            
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 7s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 7s using random strategy", "\n")
                dice7 = tuple(choice(dR7))
                state[dice7] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 7s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 7s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 7s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 7s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn 


def cells_eight_decision(state, player, diceRoll, skipped_turn, strategy):
    dR8 = np.argwhere(8 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 8 and state[1][2] == player and state[2][4] == player and state[3][1] == player and state[4][3] == player:
        text2 = font3.render("You are sitting on all Dice Face " +str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
      
    if diceRoll == 8 and state[1][2] == chip and state[2][4]  == chip and state[3][1] == chip and state[4][3] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n") 
    
    elif len(dR8) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")

    elif state[1][2] == 8 or state[2][4] == 8 or state[3][1] == 8 or state[4][3] == 8:
        if len(dR8) == 1:
            text2 = font3.render("Place chip on last 8 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 8 left", "\n")
            dice8 = tuple(choice(dR8))
            state[dice8] = player
            displayBoard(state)
            print("\n\n")
               
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 8s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 8s using random strategy", "\n")
                dice8 = tuple(choice(dR8))
                state[dice8] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 8s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 8s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 8s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 8s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn

def cells_nine_decision(state, player, diceRoll, skipped_turn, strategy):
    dR9 = np.argwhere(9 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 9 and state[1][3] == player and state[2][1] == player and state[3][4] == player and state[4][2] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
       
    if diceRoll == 9 and state[1][3] == chip and state[2][1]  == chip and state[3][4] == chip and state[4][2] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n") 
            
    elif len(dR9) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")

    elif state[1][3] == 9 or state[2][1] == 9 or state[3][4] == 9 or state[4][2] ==9:
        if len(dR9) == 1:
            text2 = font3.render("Place chip on last 9 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 9 left", "\n")
            dice9 = tuple(choice(dR9))
            state[dice9] = player
            displayBoard(state)
            print("\n\n")
               
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 9s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 9s using random strategy", "\n")
                dice9 = tuple(choice(dR9))
                state[dice9] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 9s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 9s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 9s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 9s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            
    return state, skipped_turn

def cells_twelve_decision(state, player, diceRoll, skipped_turn, strategy):
    dR12 = np.argwhere(12 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 12 and state[2][2] == player and state[2][3] == player and state[3][2] == player and state[3][3] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
      
    if diceRoll == 12 and state[2][2] == chip and state[2][3]  == chip and state[3][2] == chip and state[3][3] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n") 
    
    elif len(dR12) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")
            
    elif state[2][2] == 12 or state[2][3] == 12 or state[3][2] == 12 or state[3][3] == 12:
        if len(dR12) == 1:
            text2 = font3.render("Place chip on last 12 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 12 left", "\n")
            dice12 = tuple(choice(dR12))
            state[dice12] = player
            displayBoard(state)
            print("\n\n")
            
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 12s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 12s using random strategy", "\n")
                dice12 = tuple(choice(dR12))
                state[dice12] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 12s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 12s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 12s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 12s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
    return state, skipped_turn       
    

def cells_two_decision(state, player, diceRoll, skipped_turn, strategy):
    dR2 = np.argwhere(2 == state)
    print("Dice Face Picked =", diceRoll, "\n")
    if player == 0:
        player = 'X'
        chip= 'O'
    else:
        player = 'O'
        chip = 'X'
    if diceRoll == 2 and state[0][0] == player and state[0][5] == player and state[5][0] == player and state[5][5] == player:
        text2 = font3.render("You are sitting on all Dice Face " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
        displayBoard(state)
        skipped_turn += 1
        print("\n\n")
         
    if diceRoll == 2 and state[0][0] == chip and state[0][5]  == chip and state[5][0] == chip and state[5][5] == chip:
            text2 = font3.render("Opposition player sitting on all " + str(diceRoll)+ " on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (20, 830))
            print("Opposition player sitting on all Dice Face", diceRoll, "\n")
            skipped_turn += 1
            displayBoard(state)
            print("\n\n") 
    
    elif len(dR2) == 0:
        text2 = font4.render(str(diceRoll)+ " is all covered on the board, do not placed chips anywhere", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print(diceRoll, " is all covered on the board, do not placed chips anywhere", "\n")
        skipped_turn += 1
        displayBoard(state)
        print("\n\n")

    elif state[0][0] == 2 or state[0][5] == 2 or state[5][0] == 2 or state[5][5] == 2:
        if len(dR2) == 1:
            text2 = font3.render("Place chip on last 2 left", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (150, 830))
            print("Place chip on last 2 left", "\n")
            dice2 = tuple(choice(dR2))
            state[dice2] = player
            displayBoard(state)
            print("\n\n")
                 
        else:
            if strategy in 'random':
                text2 = font3.render("Chip placed on one of 2s using random strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 2s using random strategy", "\n")
                dice2 = tuple(choice(dR2))
                state[dice2] = player
                displayBoard(state)
                print("\n\n")
            elif strategy in 'offensive':
                text2 = font3.render("Chip placed on one of 2s using offensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 2s using offensive strategy", "\n")
                offensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
            elif strategy in 'defensive':
                text2 = font3.render("Chip placed on one of 2s using defensive strategy", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (20, 830))
                print("Chip placed on one of 2s using defensive strategy", "\n")
                defensive_strategy_two_players(state, diceRoll, player)
                displayBoard(state)
                print("\n\n")
    return state, skipped_turn       
    
        
def cells_ten_eleven_decision(state, player, diceRoll, skipped_turn):
    if diceRoll == 10:
        text2 = font4.render("Dice Rolled: "+ str(diceRoll)+ ", Do NOT Remove any chip from player", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("Dice Face Picked =", diceRoll, ", Do NOT Remove any chip from player ", "\n")
    else:
        text2 = font4.render("Dice Rolled: "+ str(diceRoll) +", Do NOT place chip anywhere on the board", 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
        screen.blit(text2, (20, 830))
        print("Dice Face Picked =", diceRoll, ", Do NOT place chip anywhere on the board", "\n")
    displayBoard(state)
    skipped_turn += 1
    print("\n\n")
    return state, skipped_turn
        
######  Is the board state or coupon Tak en/VALID?
def validity(state, player, diceRoll, skipped_turn, strategy):
    if diceRoll == 3:
        state, skipped_turn = cells_three_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 4:
        state, skipped_turn = cells_four_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 5:
        state, skipped_turn= cells_five_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 6:
        state, skipped_turn = cells_six_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 7:
        state, skipped_turn = cells_seven_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 8:
        state, skipped_turn = cells_eight_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 9:
        state, skipped_turn = cells_nine_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 12:
        state, skipped_turn = cells_twelve_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 2:
        state, skipped_turn = cells_two_decision(state, player, diceRoll, skipped_turn, strategy)
        
    elif diceRoll == 10 or diceRoll == 11:
        state, skipped_turn = cells_ten_eleven_decision(state, player, diceRoll, skipped_turn)
    return state, skipped_turn    
    

def gameResult(state, player):
    gameWon = False
    chip = ''
    if player == 0: chip = 'X'
    elif player ==1: chip = 'O'
    for row in range(len(state)):
        if tuple(state[row,1:]).count(chip)==5 or tuple(state[row,0:5]).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Row "+ str(row+1), 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))   
            break


    for column in range(len(state)):
        if tuple(state[0:5,column]).count(chip)==5 or tuple(state[1:,column]).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Column "+ str(column+1), 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))  
            break


    for diagLength in range(len(state)-5):
        if tuple(np.delete(state.diagonal(diagLength), 5)).count(chip)==5 or tuple(np.delete(state.diagonal(diagLength), 0)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, 10, 10)
            text2 = font3.render("Sequence of 5 achieved on Diagonal", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))   
            break


        if tuple(np.delete(np.fliplr(state).diagonal(diagLength),5)).count(chip)==5 or tuple(np.delete(np.fliplr(state).diagonal(diagLength),0)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Diagonal", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))  
            break


    for diagLength in range(len(state)-5):
        if tuple(state.diagonal(diagLength+1)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Diagonal Upper", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))  
            break

        elif tuple(np.fliplr(state).diagonal(diagLength+1)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Diagonal Inverse Upper", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))   
            break


    for diagLength in range(len(state)-5):
        if tuple(state.diagonal(diagLength-1)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, 10, 10)
            text2 = font3.render("Sequence of 5 achieved on Diagonal Lower", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60))  
            break

        elif tuple(np.fliplr(state).diagonal(diagLength-1)).count(chip)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, YELLOW)
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
                screen.blit(text2, (10, 10))
            text2 = font3.render("Sequence of 5 achieved on Diagonal Inverse Lower", 1, (237, 28, 36))
                #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (10, 60) )  
            break
        
    return gameWon

def draw_board(state):
    cells = [[2,3,4,5,6,2],[6,7,8,9,7,3],[5,9,12,12,8,4],[4,8,12,12,9,5],[3,7,9,8,7,6],[2,6,5,4,3,2]]
    board_cells = np.array(cells, dtype=object)
    # Draw the lines
    for i in range(6):
        for j in range(6):
            arr1 = np.swapaxes(board_cells, 0, 1)
            arr2 = np.swapaxes(state, 0, 1)
            #pygame.draw.rect(screen, GREEN, (i * square_size, j * square_size+square_size, square_size, square_size))
            pygame.draw.rect(screen, WHITE, (i * square_size, j * square_size+square_size, square_size, square_size))
                
            # Fill blue color in already numbered grid
            if arr2[i][j] == 'X':
                pygame.draw.circle(screen, GREEN, (int(i * square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)
                
           
            if arr2[i][j] == 'O':
                pygame.draw.circle(screen, YELLOW, (int(i * square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)
                
            text = font1.render(str(arr1[i][j]), 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text, (i * square_size + 30, j * square_size+square_size + 30))

            text2 = font2.render(str(arr1[i][j]), 1, (237, 28, 36))
            #screen.blit(pygame.transform.rotate(screen, 180), (0,0))
            screen.blit(text2, (i * square_size+5, j * square_size+square_size+5))
            pygame.draw.line(screen, (56, 182, 255), (0, j * square_size+square_size), (700, j * square_size+square_size), 5)
            pygame.draw.line(screen, (56, 182, 255), (j * square_size+square_size, 0), (j * square_size+square_size, 700), 5) 
   
        pygame.draw.line(screen, (56, 182, 255), (0, i * square_size), (600, i * square_size), 5)
        pygame.draw.line(screen, (56, 182, 255), (i * square_size, 0), (i * square_size, 700), 5)
    
def gamePlayTwoPlayerNoRule(strategy_one, strategy_two):
    boardCells = [[2,3,4,5,6,2],[6,7,8,9,7,3],[5,9,12,12,8,4],[4,8,12,12,9,5],[3,7,9,8,7,6],[2,6,5,4,3,2]]
    board = np.array(boardCells, dtype=object)
    pygame.init()
    pygame.display.set_caption("Sequence Dice Game - Full board Two Players Without Rules")
    draw_board(board)
    pygame.display.update()
    gameWon = False
    countOne = 0
    countTwo = 0
    turn =0
    chip_one = 20
    chip_two = 20
    skipped_turn = 0
    skipped_turn_one = 0
    skipped_turn_two = 0
    draw_count = 0
    player_one_win_count = 0
    player_two_win_count = 0
    FPS = 3000
    player_dice_face1 = None
    player_dice_face2 = None
    clock = pygame.time.Clock()
    print("Initial board display \n")
    print(displayBoard(board))
    while gameWon == False:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                gameWon == True
                pygame.quit()
                quit()
            
        if chip_one == 0 and chip_two == 0:
            text2 = font3.render("Draw Game, both players exhausted their chips", 1, (237, 28, 36))
            screen.blit(text2, (10, 30))
            pygame.display.update()
            print("Draw Game, both players exhausted their chips","\n")
            draw_count +=1
            gameWon == True
            break
        else:
            if turn == 0:
                countOne += 1
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 0
                print("Player ONE turn, game play ", countOne, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_turn_one = validity(board, player, diceRoll, skipped_turn, strategy_one)
                chip_one = chip_one -1 + skipped_chip_one
                skipped_turn_one += skipped_chip_one
                print("Player ONE has ", chip_one," chips left", "\n")
                if player_dice_face1 != None and player_dice_face2 != None:
                    text2 = font2.render('Player one turns, game play: '+str(countOne), 1, (237, 28, 36))
                    screen.blit(text2, (100, 710))
                    screen.blit(player_dice_face1, (180, 750))
                    screen.blit(player_dice_face2, (240, 750))
                    text3= font2.render('chips left: ' +str(chip_one), 1, (237, 28, 36))
                    screen.blit(text3, (450, 710))
                    pygame.display.update()
                    pygame.time.delay(2000)
                gameWon = gameResult(move, player)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(2000)
                for i in range(6):
                    pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size,800, square_size, square_size))
                    pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size,0, square_size, square_size))
                pygame.display.update()
                for i in range(6):
                        for j in range(6):
                            pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size, j+7*square_size, square_size, square_size))
                pygame.display.update()
                if gameWon == True:
                    player_one_win_count +=1
                    break
                else:
                    gameWon = False
            else:
                countTwo += 1
                pygame.display.update()
                pygame.time.delay(2000)
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 1
                print("Player TWO turn, game play ", countTwo, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_turn = validity(board, player, diceRoll ,skipped_turn, strategy_two)
                chip_two = chip_two -1 + skipped_chip_two
                skipped_turn_two += skipped_chip_two
                print("Player TWO has ", chip_two-1," chips left", "\n")
                if player_dice_face1 != None and player_dice_face2 != None:
                    text3= font2.render('Player two turns, game play: '+str(countTwo), 1, (237, 28, 36))
                    screen.blit(player_dice_face1, (180, 750))
                    screen.blit(player_dice_face2, (240, 750))
                    screen.blit(text3, (100, 710))
                    text3= font2.render('chips left: ' +str(chip_two), 1, (237, 28, 36))
                    screen.blit(text3, (450, 710))
                    pygame.display.update()
                    pygame.time.delay(3000)
                gameWon = gameResult(move, player)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(3000)
                for i in range(6):
                    pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size,800, square_size, square_size))
                pygame.display.update()
                for i in range(6):
                        for j in range(6):
                            pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size, j+7*square_size, square_size, square_size))
                            pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size,0, square_size, square_size))
                pygame.display.update()
                if gameWon == True:
                    player_two_win_count +=1
                    break
                else:
                    gameWon = False

        turn += 1
        turn = turn % 2
    print("Player one game length: " + str(countOne))
    print("Player two game length: " + str(countTwo))
    print("Player One win Count: " + str(player_one_win_count))
    print("Player Two win Count: " + str(player_two_win_count))
    print("Player one skipped turns length: " + str(skipped_turn_one))
    print("Player two skipped turns length: " + str(skipped_turn_two))
    return countOne, countTwo,skipped_turn_one,skipped_turn_two,player_one_win_count, player_two_win_count, draw_count
    pygame.quit()
    quit()
    

        
if __name__=='__main__':
    strategy_one = input("Welcome Player One, What strategy do you want to use: e.g offensive defensive, random: \n").lower()
    strategy_two = input("Welcome Player Two, What strategy do you want to use: e.g offensive, defensive, random: \n").lower()
    gamePlayTwoPlayerNoRule(strategy_one, strategy_two)
```
