```python

from random import randint, choice
import numpy as np
import pygame, sys
from pygame.locals import *
import pygame.surfarray as surfarray
square_size = 100
column_size = 6
row_size = 6
WHITE = (255, 255, 255)
LIGHT_BLUE = (56, 182, 255)
GREEN = (0,255,0)
YELLOW = (255,255,0)
square_size = 100
RADIUS = int(square_size/2 -10)
width = column_size * square_size
height = (row_size+1) * square_size
size = (width, height)
pygame.font.init()
 
# Total window
screen = pygame.display.set_mode((1100, 400))
screen.fill(LIGHT_BLUE)
val = 0
# Load test fonts for future use
font1 = pygame.font.SysFont("comicsans", 40)
font2 = pygame.font.SysFont("comicsans", 20)
font3 = pygame.font.SysFont("Calibri", 25)
font4 = pygame.font.SysFont("comicsans", 20)

dice1 = pygame.image.load("dice_1.png")
dice2 = pygame.image.load("dice_2.png")
dice3 = pygame.image.load("dice_3.png")
dice4 = pygame.image.load("dice_4.png")
dice5 = pygame.image.load("dice_5.png")
dice6 = pygame.image.load("dice_6.png")
dice_face = [None, dice1, dice2, dice3, dice4, dice5, dice6]

def displayOneDimBoard(state):
    print("[",state[0]," |",state[1], " |",state[2], " |",state[3]," |",state[4]," |",state[5]," |",state[6]," |",
          state[7]," |",state[8]," ]","\n\n")
    

def validity(state, player, diceRoll,skipped_turn):
    print("Dice Face Picked =", diceRoll, "\n")
    if (diceRoll != 10 and diceRoll != 11) and tuple(state).count(diceRoll) == 0:
        text2 = font3.render("No "+str(diceRoll)+" left on the board Don't Place chip anywhere", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No ",diceRoll, " left on the board Don't Place chip anywhere", "\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
            
    else:
        for number in range(len(state)):
            if state[number] == diceRoll:
                if player == 0:
                    state[number] = 'X' 
                else:
                    state[number] = 'O'
                displayOneDimBoard(state)
                print("\n\n")
                break

            
    if diceRoll == 10 or diceRoll == 11:
        if diceRoll == 10:
            text2 = font3.render("Do NOT Remove the opponent's chip", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("Dice Face Picked =", diceRoll, "Do NOT Remove any chip the opponent", "\n")
        else:
            text2 = font3.render("Do NOT place chip anywhere on the board", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("Dice Face Picked =", diceRoll, "Do NOT place chip anywhere on the baord", "\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
    
    return state,skipped_turn

        
def gameResult(state,player):
    if player == 0: player = 'X'
    else: player = 'O'
    gameWon = False
    for row in range(len(state)):
        if tuple(state[row:5+row]).count(player)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                screen.blit(text2, (200, 10))
                print("GAME - OVER: PLAYER ", 1, " WINS, CONGRATULATIONS!!!","\n")
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, GREEN)
                screen.blit(text2, (200, 10))
                print("GAME - OVER: PLAYER ", 2, " WINS, CONGRATULATIONS!!!","\n")
            text2 = font3.render("Sequence of 5 achieved", 1, (237, 28, 36))
            screen.blit(text2, (200, 60)) 
            print("Sequence of 5 achieved ","\n")
            break
    return gameWon

def draw_board(state):
    board_cells = np.array([2,3,4,5,6,7,8,9,12], dtype=object)
    for i in range(9):
        for j in range(1):
            pygame.draw.rect(screen, WHITE, (i * square_size+square_size, j * square_size+square_size, square_size, square_size))
            if state[i] == 'X':
                pygame.draw.circle(screen, GREEN, (int(i * square_size+square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)
            if state[i] == 'O':
                pygame.draw.circle(screen, YELLOW, (int(i * square_size+square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)

            text = font1.render(str(board_cells[i]), 1, (237, 28, 36))
            screen.blit(text, (i * square_size+square_size + 30, j * square_size+square_size + 30))

            text2 = font2.render(str(board_cells[i]), 1, (237, 28, 36))
            screen.blit(text2, (i * square_size+square_size+5, j * square_size+square_size+5))
        pygame.draw.line(screen, (56, 182, 255), (i * square_size+square_size, 100), (i * square_size+square_size, 200), 5)


def gamePlayTwoPlayerOneDimensionNoRule():
    board = np.array([2,3,4,5,6,7,8,9,12], dtype=object)
    pygame.init()
    pygame.display.set_caption("Sequence Dice Game - One Dimension Two Players Without Rules")
    draw_board(board)
    pygame.display.update()
    gameWon = False
    countOne = 0
    countTwo = 0
    turn =0
    chip_one = 20
    chip_two = 20
    skipped_turn = 0
    skipped_turn_one = 0
    skipped_turn_two = 0
    draw_count = 0
    player_dice_face1 = None
    player_dice_face2 = None
    clock = pygame.time.Clock()
    player_one_win_count = 0
    player_two_win_count = 0
    print("Initial board display \n")
    print(displayOneDimBoard(board))
    while gameWon == False:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                gameWon == True
                pygame.quit()
                quit()
            
        if chip_one == 0 and chip_two == 0:
            text2 = font3.render("Draw Game, both players exhausted their chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 30))
            pygame.display.update()
            print("Draw Game, both players exhausted their chips","\n")
            draw_count +=1
            gameWon == True
            break
        else:
            if turn == 0:
                countOne += 1
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 0
                print("Player ONE turn, game play ", countOne, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_chip_one = validity(board, player, diceRoll, skipped_turn)
                gameWon = gameResult(move,player)
                chip_one = chip_one -1 + skipped_chip_one
                print("Player ONE has ", chip_one," chips left", "\n")
                skipped_turn_one += skipped_chip_one
                if player_dice_face1 != None and player_dice_face2 != None:
                    text2 = font2.render('Player one turns, game play: '+str(countOne), 1, (237, 28, 36))
                    screen.blit(text2, (200, 210))
                    screen.blit(player_dice_face1, (280, 250))
                    screen.blit(player_dice_face2, (340, 250))
                    text3= font2.render('chips left: ' +str(chip_one), 1, (237, 28, 36))
                    screen.blit(text3, (850, 210))
                    pygame.display.update()
                    pygame.time.delay(2000)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(3000)
                for i in range(9):
                    for j in range(1):
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size,300, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size,0, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size, j+2*square_size, square_size, square_size))
                pygame.display.update()
                if gameWon == True:
                    player_one_win_count +=1
                    break
                else:
                    gameWon = False
            else:
                countTwo += 1
                pygame.display.update()
                pygame.time.delay(3000)
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 1
                print("Player TWO turn, game play ", countTwo, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_chip_two = validity(board, player, diceRoll ,skipped_turn)
                gameWon = gameResult(move, player)
                chip_two = chip_two - 1 + skipped_chip_two
                print("Player TWO has ", chip_two," chips left", "\n")
                skipped_turn_two += skipped_chip_two
                if player_dice_face1 != None and player_dice_face2 != None:
                    text3= font2.render('Player two turns, game play: '+str(countTwo), 1, (237, 28, 36))
                    screen.blit(player_dice_face1, (280, 250))
                    screen.blit(player_dice_face2, (340, 250))
                    screen.blit(text3, (200, 210))
                    text3= font2.render('chips left: ' +str(chip_two), 1, (237, 28, 36))
                    screen.blit(text3, (850, 210))
                    pygame.display.update()
                    pygame.time.delay(3000)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(3000)
                for i in range(9):
                    for j in range(1):
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size,300, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size,0, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size, j+2*square_size, square_size, square_size))
                pygame.display.update()

                if gameWon == True:
                    player_two_win_count +=1
                    break
                else:
                    gameWon = False

        turn += 1
        turn = turn % 2
    print("Player one game length: " + str(countOne))
    print("Player two game length: " + str(countTwo))
    print("Player One win Count: " + str(player_one_win_count))
    print("Player Two win Count: " + str(player_two_win_count))
    print("Player one skipped turns length: " + str(skipped_turn_one))
    print("Player two skipped turns length: " + str(skipped_turn_two))
    return countOne, countTwo,skipped_turn_one,skipped_turn_two, player_one_win_count, player_two_win_count, draw_count

    pygame.quit()
    quit()


if __name__=='__main__':
    gamePlayTwoPlayerOneDimensionNoRule()
    pygame.quit()
    quit()
```

```python
#Run for 1000 simulations
```

```python
import statistics
import random
player_one_win_length =[]
player_two_win_length =[]
player_one_game_count =[]
player_two_game_count =[]
skipped_turn_length = []
total_game_length = []
draw_length = []
random.seed(22)
for item in range(1000):
    countOne, countTwo, skipped_turn_one,skipped_turn_two, player_one_win_count, player_two_win_count,draw_count = gamePlayTwoPlayerOneDimensionNoRule()
    print("number of plays: " + str(item+1))
    player_one_win_length.append(player_one_win_count)
    player_two_win_length.append(player_two_win_count)
    player_one_game_count.append(countOne)
    player_two_game_count.append(countTwo)
    player_one_skipped_turn.append(skipped_turn_one)
    player_two_skipped_turn.append(skipped_turn_two)
    skipped_turn_length.append(skipped_turn)
    draw_length.append(draw_count)
    total_skipped_turn.append(skipped_turn_one + skipped_turn_two)
    total_game_length.append(countOne + countTwo)
    averageGameLength = statistics.mean(total_game_length)
    averageSkippedLength = statistics.mean(total_skipped_turn)
print(f" skipped turns length : {skipped_turn_length}")
print(f" draw game count : {draw_length}")
print(f" player one game counts : {player_one_game_count}")
print(f" player two game counts : {player_one_game_count}")
print(f" player one win lengths : {player_one_win_length}")
print(f" player two win lengths : {player_two_win_length}")
print(f" Total game lengths : {total_skipped_turn}")
print(f" Total skipped turns : {total_game_length}")
print(f" Total average game lengths : {round(averageGameLength, 0)}")
print(f" Total skipped turn lengths : {round(averageSkippedLength, 0)}")
print(f" Total player one win in 1000 play : {sum(player_one_win_length)}") 
print(f" Total player two win in 1000 play : {sum(player_two_win_length)}")
print(f" Total player one skipped turn in 1000 play : {sum(player_one_skipped_turn)}") 
print(f" Total player two skipped turn in 10 play : {sum(player_two_skipped_turn)}") 
print(f" Total draw in 100 play : {sum(draw_length)}") 
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.title("Histogram of Total Game Length")
plt.xlabel("Number of Rolls")
plt.ylabel("Probability")
mu, sd = norm.fit(total_game_length)
md = statistics.mode(total_game_length)
mid = statistics.median(total_game_length)
plt.hist(total_game_length, density = True, color='plum', edgecolor = 'black')
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax, 100)
p = norm.pdf(x, mu, sd)
plt.plot(x, p,'k', linewidth=2)
plt.axvline(mu, color='red',linestyle='dashed', linewidth=2)
plt.axvline(md, color='b', linewidth=2)
plt.axvline(mid, color='k', linewidth=2)
plt.legend(['Median', 'Mean','Mode'])
plt.show()


plt.title("Histogram of Skipped Turn Length")
plt.xlabel("Number of Rolls")
plt.ylabel("Probability")
mu1, sd1 = norm.fit(skipped_turn_length)
md1 = statistics.mode(skipped_turn_length)
mid1 = statistics.median(skipped_turn_length)
plt.hist(skipped_turn_length, density = True, color='plum', edgecolor = 'black')
xmin1, xmax1 = plt.xlim()
x1 = np.linspace(xmin1,xmax1, 100)
p1 = norm.pdf(x1, mu1, sd1)
plt.plot(x1, p1,'k', linewidth=2)
plt.axvline(mu1, color='red',linestyle='dashed', linewidth=2)
plt.axvline(md1, color='b', linewidth=2)
plt.axvline(mid1, color='k', linewidth=2)
plt.legend(['Median', 'Mean','Mode'])
plt.show()


plt.title("Histogram of Player One Game Length")
plt.xlabel("Number of Rolls")
plt.ylabel("Probability")
mu2, sd2 = norm.fit(player_one_game_count)
md2 = statistics.mode(player_one_game_count)
mid2 = statistics.median(player_one_game_count)
plt.hist(player_one_game_count, density = True, color='plum', edgecolor = 'black')
xmin2, xmax2 = plt.xlim()
x2 = np.linspace(xmin1,xmax1, 100)
p2 = norm.pdf(x2, mu2, sd2)
plt.plot(x2, p2,'k', linewidth=2)
plt.axvline(mu2, color='red',linestyle='dashed', linewidth=2)
plt.axvline(md2, color='b', linewidth=2)
plt.axvline(mid2, color='k', linewidth=2)
plt.legend(['Median', 'Mean','Mode'])
plt.show()


plt.title("Histogram of Player tow Game Length")
plt.xlabel("Number of Rolls")
plt.ylabel("Probability")
mu3, sd3 = norm.fit(player_two_game_count)
md3 = statistics.mode(player_two_game_count)
mid3 = statistics.median(player_two_game_count)
plt.hist(player_two_game_count, density = True, color='plum', edgecolor = 'black')
xmin3, xmax3 = plt.xlim()
x3 = np.linspace(xmin1,xmax1, 100)
p3 = norm.pdf(x3, mu3, sd3)
plt.plot(x3, p3,'k', linewidth=2)
plt.axvline(mu3, color='red',linestyle='dashed', linewidth=2)
plt.axvline(md3, color='b', linewidth=2)
plt.axvline(mid3, color='k', linewidth=2)
plt.legend(['Median', 'Mean','Mode'])
plt.show()


plt.title("Boxplot Distribution of Total Length of Game And Skipped Turn Length")
plt.ylabel("Number of Rolls")
plt.boxplot([total_game_length, skipped_turn_length], patch_artist=True)
plt.xticks([1,2], ['TGL','STL'])
plt.legend(['Total Game Length', 'Skipped Turn Length'],loc="center")
plt.show()

plt.title("Boxplot Distribution of Length of Game For player One and Two")
plt.ylabel("Number of Rolls")
plt.boxplot([player_one_game_count, player_two_game_count], patch_artist=True)
plt.xticks([1,2], ['POGC', 'PTGC'])
plt.legend(['Player One Game Count', 'Player Two Game Count'], loc="center")
plt.show()
```
