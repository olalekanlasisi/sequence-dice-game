```python
from random import randint, choice
import numpy as np
import pygame, sys
from pygame.locals import *
import pygame.surfarray as surfarray
square_size = 100
column_size = 6
row_size = 6
WHITE = (255, 255, 255)
LIGHT_BLUE = (56, 182, 255)
GREEN = (0,255,0)
YELLOW = (255,255,0)
square_size = 100
RADIUS = int(square_size/2 -10)
width = column_size * square_size
height = (row_size+1) * square_size
size = (width, height)
pygame.font.init()
 
# Total window
screen = pygame.display.set_mode((1100, 400))
screen.fill(LIGHT_BLUE)
val = 0
# Load test fonts for future use
font1 = pygame.font.SysFont("comicsans", 40)
font2 = pygame.font.SysFont("comicsans", 20)
font3 = pygame.font.SysFont("Calibri", 25)
font4 = pygame.font.SysFont("comicsans", 20)

dice1 = pygame.image.load("dice_1.png")
dice2 = pygame.image.load("dice_2.png")
dice3 = pygame.image.load("dice_3.png")
dice4 = pygame.image.load("dice_4.png")
dice5 = pygame.image.load("dice_5.png")
dice6 = pygame.image.load("dice_6.png")
dice_face = [None, dice1, dice2, dice3, dice4, dice5, dice6]


def displayOneDimBoard(state):
    print("[",state[0]," |",state[1], " |",state[2], " |",state[3]," |",state[4]," |",state[5]," |",state[6]," |",
          state[7]," |",state[8]," ]","\n\n")
    
def ruleTenLowestProbability(state, diceRoll, player,skipped_turn):
    sliceList =[]
    probability = np.array([0.027, 0.05, 0.083, 0.1, 0.138, 0.16, 0.138, 0.1, 0.027])
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    if player == 0: player = 'O'
    else: player = 'X'
    zero = np.where(player == state)
    if(len(zero[0]) == 0):
        text2 = font3.render("No chips available for Removal", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    elif len(zero[0]) == 1 and (state[0]== player or state[8] == player ):
        if state[0]== player:
            text2 = font3.render("No chip(s) available for Removal: chip(s) is on 2", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 2","\n")
        else:
            ext2 = font3.render("No chip(s) available for Removal: chip(s) is on 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    
    elif len(zero[0]) == 2 and (state[0]== player and state[8] == player):
        text2 = font3.render("No chips available for Removal: chips are on 2 and 12", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal: chips are on 2 and 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")

      
    else:
        if state[0]== player and state[8] == player:
            text2 = font3.render("chips can not be removed from cell 2 and 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("chips can not be removed from cell 2 and 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-2)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-2, " cells covered by opponent's chips")
        elif state[0]== player or state[8] == player:
            if state[0]== player:
                text2 = font3.render("chips can not be removed from cell 2", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 2","\n")
            else:
                text2 = font3.render("chips can not be removed from cell 12", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-1)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-1, " cells covered by opponent's chips")
        else:
            text2 = font3.render("There are other "+str(len(zero[0]))+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("There are ", len(zero[0]), " cells covered by opponent's chips")

        chip = np.where(player == state[np.array([1, 2, 3, 4,5,6,7])])
        for i in range(len(chip[0])):
            B = chip[0][i]
            sliceList.append(probability[B+1])
        for i in range(len(chip[0])):
            B = chip[0][i]
            if(np.amin(sliceList) == probability[B+1]):
                state[B+1] =   mainCell[B+1]
                text2 = font3.render("Remove opposition chip from "+str(mainCell[B+1])+" which has lowest probability", 1, (237, 28, 36))
                screen.blit(text2, (200, 370))
                print("Remove opposition chip from ",  mainCell[B+1], " which has lowest probability", "\n")
                displayOneDimBoard(state)
                print("\n\n")
                break
    
    return state

def ruleTenHighestProbability(state,diceRoll,player, skipped_turn):
    sliceList =[]
    probability = np.array([0.027, 0.05, 0.083, 0.1, 0.138, 0.16, 0.138, 0.1, 0.027])
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    if player == 0: player = 'O'
    else: player = 'X'
    zero = np.where(player == state)
    if(len(zero[0]) == 0):
        text2 = font3.render("No chips available for Removal", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    elif len(zero[0]) == 1 and (state[0]== player or state[8] == player ):
        if state[0]== player:
            text2 = font3.render("No chip(s) available for Removal: chip(s) is on 2", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 2","\n")
        else:
            ext2 = font3.render("No chip(s) available for Removal: chip(s) is on 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    
    elif len(zero[0]) == 2 and (state[0]== player and state[8] == player):
        text2 = font3.render("No chips available for Removal: chips are on 2 and 12", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal: chips are on 2 and 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")

      
    else:
        if state[0]== player and state[8] == player:
            text2 = font3.render("chips can not be removed from cell 2 and 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("chips can not be removed from cell 2 and 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-2)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-2, " cells covered by opponent's chips")
        elif state[0]== player or state[8] == player:
            if state[0]== player:
                text2 = font3.render("chips can not be removed from cell 2", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 2","\n")
            else:
                text2 = font3.render("chips can not be removed from cell 12", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-1)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-1, " cells covered by opponent's chips")
        else:
            text2 = font3.render("There are other "+str(len(zero[0]))+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("There are ", len(zero[0]), " cells covered by opponent's chips")

        chip = np.where(player == state[np.array([1, 2, 3, 4,5,6,7])])
        for i in range(len(chip[0])):
            B = chip[0][i]
            sliceList.append(probability[B+1])
        for i in range(len(chip[0])):
            B = chip[0][i]
            if(np.amax(sliceList) == probability[B+1]):
                state[B+1] =   mainCell[B+1]
                text2 = font3.render("Remove opposition chip from "+str(mainCell[B+1])+" which has highest probability", 1, (237, 28, 36))
                screen.blit(text2, (200, 370))
                print("Remove opposition chip from ",  mainCell[B+1], " which has highest probability", "\n")
                displayOneDimBoard(state)
                print("\n\n")
                break
    
    return state

def ruleTenRandom(state, diceRoll,player, skipped_turn):
    sliceList =[]
    probability = np.array([0.027, 0.05, 0.083, 0.1, 0.138, 0.16, 0.138, 0.1, 0.027])
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    if player == 0: player = 'O'
    else: player = 'X'
    zero = np.where(player == state)
    if(len(zero[0]) == 0):
        text2 = font3.render("No chips available for Removal", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    elif len(zero[0]) == 1 and (state[0]== player or state[8] == player ):
        if state[0]== player:
            text2 = font3.render("No chip(s) available for Removal: chip(s) is on 2", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 2","\n")
        else:
            text2 = font3.render("No chip(s) available for Removal: chip(s) is on 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("No chip(s) available for Removal: chip(s) is on 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")
        
    
    elif len(zero[0]) == 2 and (state[0]== player and state[8] == player):
        text2 = font3.render("No chips available for Removal: chips are on 2 and 12", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("No chips available for Removal: chips are on 2 and 12","\n")
        displayOneDimBoard(state)
        skipped_turn +=1
        print("\n\n")

      
    else:
        if state[0]== player and state[8] == player:
            text2 = font3.render("chips can not be removed from cell 2 and 12", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("chips can not be removed from cell 2 and 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-2)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-2, " cells covered by opponent's chips")
        elif state[0]== player or state[8] == player:
            if state[0]== player:
                text2 = font3.render("chips can not be removed from cell 2", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 2","\n")
            else:
                text2 = font3.render("chips can not be removed from cell 12", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("chips can not be removed from cell 12","\n")
            text2 = font3.render("There are other "+str(len(zero[0])-1)+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 360))
            print("There are other", len(zero[0])-1, " cells covered by opponent's chips")
        else:
            text2 = font3.render("There are other "+str(len(zero[0]))+" cells covered by opponent's chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 330))
            print("There are ", len(zero[0]), " cells covered by opponent's chips")

        chip = np.where(player == state[np.array([1, 2, 3, 4,5,6,7])])
        dice10 = choice(chip)
        B = dice10[0]
        state[B+1] =  mainCell[B+1]
        text2 = font3.render("Remove opposition chip from "+str(mainCell[B+1])+" Randomly", 1, (237, 28, 36))
        screen.blit(text2, (200, 370))
        print("Remove opposition chip from ", mainCell[B+1], " Randomly", "\n")
        displayOneDimBoard(state)
        print("\n\n")
       
    return state
    
def ruleElevenLowestProbability(state, diceRoll, player):
    sliceList =[]
    probability = np.array([0.027, 0.05, 0.083, 0.1, 0.138, 0.16, 0.138, 0.1, 0.027])
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    # print("Dice Face Picked =", diceRoll, "Do NOT Remove any chip from PLAYER ", "\n")
   
    zero =  np.where(np.logical_and(state != 'X',state != 'O' ))
    if len(zero[0]) > 0:
        text2 = font3.render("There are "+str(len(zero[0]))+" cells yet to be covered by chips", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("There are ", len(zero[0]), " cells yet to be covered by opponent's chips")
        for i in range(len(zero[0])):
            B = zero[0][i]
            sliceList.append(probability[B])
        for i in range(len(zero[0])):
            B = zero[0][i]
            if(np.amin(tuple(sliceList)) == probability[B]):
                if player == 0:
                    state[B] = 'X'
                else:
                    state[B] = 'O'
                text2 = font3.render("Chip placed on cell "+str(mainCell[B])+" which has lowest probability", 1, (237, 28, 36))
                screen.blit(text2, (200, 360))
                print("Chip placed on cell ", mainCell[B], " which has lowest probability", "\n")
                displayOneDimBoard(state)
                print("\n\n")
                break
    return state


def ruleElevenHighestProbability(state,diceRoll, player):
    sliceList =[]
    probability = np.array([0.027, 0.05, 0.083, 0.1, 0.138, 0.16, 0.138, 0.1, 0.027])
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    # print("Dice Face Picked =", diceRoll, "Do NOT Remove any chip from PLAYER ", "\n")
    zero =  np.where(np.logical_and(state != 'X',state != 'O' ))
    if len(zero[0]) > 0:
        text2 = font3.render("There are "+str(len(zero[0]))+" cells yet to be covered by chips", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("There are ", len(zero[0]), " cells yet to be covered by opponent's chips")
        for i in range(len(zero[0])):
            B = zero[0][i]
            sliceList.append(probability[B])
        for i in range(len(zero[0])):
            B = zero[0][i]
            if(np.amax(sliceList) == probability[B]):
                if player == 0:
                    state[B] = 'X'
                else:
                    state[B] = 'O'
                text2 = font3.render("Chip placed on cell "+str(mainCell[B])+" which has highest probability", 1, (237, 28, 36))
                screen.blit(text2, (200, 360))
                print("Chip placed on cell ", mainCell[B], "which has highest probability","\n")
                displayOneDimBoard(state)
                print("\n\n")
                break
    return state

def ruleElevenRandom(state, diceRoll, player):
    sliceList =[]
    mainCell = np.array([2,3,4,5,6,7,8,9,12])
    # print("Dice Face Picked =", diceRoll, "Do NOT Remove any chip from PLAYER ", "\n")
    zero =  np.where(np.logical_and(state != 'X',state != 'O' ))
    if len(zero[0]) > 0:
        text2 = font3.render("There are "+str(len(zero[0]))+" cells yet to be covered by chips", 1, (237, 28, 36))
        screen.blit(text2, (200, 330))
        print("There are ", len(zero[0]), " cells yet to be covered by chips")
        dice10 = choice(zero)
        B = dice10[0]
        if player == 0:
            state[B] = 'X'
        else:
            state[B] = 'O'
        text2 = font3.render("Chip placed on cell "+str(mainCell[B])+" Randomly", 1, (237, 28, 36))
        screen.blit(text2, (200, 360))
        print("Chip Placed on cell ", mainCell[B], " Randomly", "\n")
        displayOneDimBoard(state)
    return state

    
def validity(state, player, diceRoll,skipped_turn, turn, ruleTenDecision, ruleElevenDecision):
    if player == 0:
        chip = 'X'
        opponent_chip ='O'
    else: 
        chip = 'O'
        opponent_chip ='X'
    print("Dice Face Picked =", diceRoll, "\n")
    for number in range(len(state)):
        mainCell = np.array([2,3,4,5,6,7,8,9,12])
        if diceRoll != 2 and diceRoll != 12  and diceRoll !=10 and diceRoll != 11:
            if state[number] == chip and mainCell[number] == diceRoll:
                text2 = font3.render("You are sitting on all "+str(diceRoll)+" on the board", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("You are sitting on all Dice Face ", diceRoll, " on the board", "\n")
                displayOneDimBoard(state)
                skipped_turn +=1
                print("\n\n")
                break
                
            elif state[number] == opponent_chip and mainCell[number] == diceRoll:
                state[number] = chip
                text2 = font3.render("Remove chips of opposition player and place your chip on the "+str(diceRoll)+" of opposing player", 1, (237, 28, 36))
                screen.blit(text2, (100, 330))
                print("Remove chips of opposition player and place your chip on the ", diceRoll," of opposing player", "\n")
                displayOneDimBoard(state)
                print("\n\n")
                break

            elif state[number] == diceRoll:
                state[number] = chip
                displayOneDimBoard(state)
                print("\n\n")
                break
                
        elif diceRoll == 2 or diceRoll == 12:
            turn = player - 1
            if state[number] == chip and mainCell[number] == diceRoll:
                text2 = font3.render("You are sitting on all "+str(diceRoll)+" on the board", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("You are sitting on all Dice Face", diceRoll, " on the board", "\n")
                text2 = font3.render("PLAYER "+str(player+1)+", roll again after placing your chip", 1, (237, 28, 36))
                screen.blit(text2, (200, 360))
                displayOneDimBoard(state)
                print("PLAYER ", player+1, ", roll again after placing your chip", "\n")
                skipped_turn +=1
                print("\n\n")
                break
                
            elif state[number] == opponent_chip and mainCell[number] == diceRoll:
                state[number] = chip
                text2 = font3.render("Remove chips of opposition player and place your chip on the "+str(diceRoll)+" of opposing player", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("Remove chips of opposition player and place your chip on the ", diceRoll," of opposing player", "\n")
                text2 = font3.render("PLAYER "+str(player+1)+", roll again after placing your chip", 1, (237, 28, 36))
                screen.blit(text2, (100, 360))
                displayOneDimBoard(state)
                print("PLAYER ", player+1, ", roll again after placing your chip", "\n")
                print("\n\n")
                break
                
            elif state[number] == diceRoll:
                state[number] = chip
                displayOneDimBoard(state)
                text2 = font3.render("PLAYER "+str(player+1)+", roll again after placing your chip", 1, (237, 28, 36))
                screen.blit(text2, (200, 330))
                print("PLAYER ", player+1, ", roll again after placing your chip", "\n")
                print("\n\n")
                break
            
                   
    if diceRoll == 10:
        if ruleTenDecision in 'lowest probability':
            state = ruleTenLowestProbability(state, diceRoll,player,skipped_turn)
            
        if ruleTenDecision in 'highest probability':
            state = ruleTenHighestProbability(state, diceRoll,player,skipped_turn)
            
        if ruleTenDecision in 'random':
            state = ruleTenRandom(state, diceRoll, player,skipped_turn)
            
    elif diceRoll == 11:
        if ruleElevenDecision in 'lowest probability':
            state = ruleElevenLowestProbability(state, diceRoll,player)
            
        if ruleElevenDecision in 'highest probability':
            state= ruleElevenHighestProbability(state, diceRoll,player)
        
        if ruleElevenDecision in 'random':
            state= ruleElevenRandom(state, diceRoll, player)
    return state,skipped_turn,turn
        
def gameResult(state,player):
    if player == 0: player = 'X'
    else: player = 'O'
    gameWon = False
    for row in range(len(state)):
        if tuple(state[row:5+row]).count(player)==5:
            gameWon = True
            if player == 0:
                text2 = font3.render("GAME - OVER: PLAYER 1 WINS, CONGRATULATIONS!!!", 1, GREEN)
                screen.blit(text2, (200, 10))
                print("GAME - OVER: PLAYER ", 1, " WINS, CONGRATULATIONS!!!","\n")
            else:
                text2 = font3.render("GAME - OVER: PLAYER 2 WINS, CONGRATULATIONS!!!", 1, GREEN)
                screen.blit(text2, (200, 10))
                print("GAME - OVER: PLAYER ", 2, " WINS, CONGRATULATIONS!!!","\n")
            text2 = font3.render("Sequence of 5 achieved", 1, (237, 28, 36))
            screen.blit(text2, (200, 60)) 
            print("Sequence of 5 achieved ","\n")
            break
    return gameWon

def draw_board(state):
    board_cells = np.array([2,3,4,5,6,7,8,9,12], dtype=object)
    for i in range(9):
        for j in range(1):
            pygame.draw.rect(screen, WHITE, (i * square_size+square_size, j * square_size+square_size, square_size, square_size))
            if state[i] == 'X':
                pygame.draw.circle(screen, GREEN, (int(i * square_size+square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)
            if state[i] == 'O':
                pygame.draw.circle(screen, YELLOW, (int(i * square_size+square_size+square_size/2), int(j * square_size+square_size+square_size/2)), RADIUS)

            text = font1.render(str(board_cells[i]), 1, (237, 28, 36))
            screen.blit(text, (i * square_size+square_size + 30, j * square_size+square_size + 30))

            text2 = font2.render(str(board_cells[i]), 1, (237, 28, 36))
            screen.blit(text2, (i * square_size+square_size+5, j * square_size+square_size+5))
        pygame.draw.line(screen, (56, 182, 255), (i * square_size+square_size, 100), (i * square_size+square_size, 200), 5)


def gamePlayTwoPlayerOneDimensionRule(player_one_rule_ten_decision, player_one_rule_eleven_decision,player_two_rule_ten_decision,player_two_rule_eleven_decision):
    board = np.array([2,3,4,5,6,7,8,9,12], dtype=object)
    pygame.init()
    pygame.display.set_caption("Sequence Dice Game - One Dimension Two Players With Rules")
    draw_board(board)
    pygame.display.update()
    gameWon = False
    countOne = 0
    countTwo = 0
    turn =0
    chip_one = 20
    chip_two = 20
    skipped_turn = 0
    skipped_turn_one = 0
    skipped_turn_two = 0
    draw_count = 0
    player_dice_face1 = None
    player_dice_face2 = None
    clock = pygame.time.Clock()
    player_one_win_count = 0
    player_two_win_count = 0
    print("Initial board display \n")
    print(displayOneDimBoard(board))
    while gameWon == False:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                gameWon == True
                pygame.quit()
                quit()
            
        if chip_one == 0 and chip_two == 0:
            text2 = font3.render("Draw Game, both players exhausted their chips", 1, (237, 28, 36))
            screen.blit(text2, (200, 30))
            pygame.display.update()
            print("Draw Game, both players exhausted their chips","\n")
            draw_count +=1
            gameWon == True
            break
        else:
            if turn == 0:
                countOne += 1
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 0
                print("Player ONE turn, game play ", countOne, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_chip_one,turn = validity(board, player, diceRoll, skipped_turn,turn,player_one_rule_ten_decision, player_one_rule_eleven_decision)
                gameWon = gameResult(move,player)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(3000)
                chip_one = chip_one -1 + skipped_chip_one
                print("Player ONE has ", chip_one," chips left", "\n")
                skipped_turn_one += skipped_chip_one
                if player_dice_face1 != None and player_dice_face2 != None:
                    text2 = font2.render('Player one turns, game play: '+str(countOne), 1, (237, 28, 36))
                    screen.blit(text2, (200, 210))
                    screen.blit(player_dice_face1, (280, 250))
                    screen.blit(player_dice_face2, (340, 250))
                    text3= font2.render('chips left: ' +str(chip_one), 1, (237, 28, 36))
                    screen.blit(text3, (850, 210))
                    pygame.display.update()
                    pygame.time.delay(2000)
                for i in range(9):
                    for j in range(1):
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size,300, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size,0, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size, j+2*square_size, square_size, square_size))
                pygame.display.update()
                if gameWon == True:
                    player_one_win_count +=1
                    break
                else:
                    gameWon = False
            else:
                countTwo += 1
                pygame.display.update()
                pygame.time.delay(3000)
                dice_roll1 = randint(1, 6)
                dice_roll2 = randint(1, 6)
                player_dice_face1  = dice_face[dice_roll1]
                player_dice_face2 = dice_face[dice_roll2]
                player = 1
                print("Player TWO turn, game play ", countTwo, "\n")
                diceRoll = sum([dice_roll1, dice_roll2])
                move, skipped_chip_two,turn = validity(board, player, diceRoll,skipped_turn,turn,player_two_rule_ten_decision,player_two_rule_eleven_decision)
                if player_dice_face1 != None and player_dice_face2 != None:
                    text3= font2.render('Player two turns, game play: '+str(countTwo), 1, (237, 28, 36))
                    screen.blit(player_dice_face1, (280, 250))
                    screen.blit(player_dice_face2, (340, 250))
                    screen.blit(text3, (200, 210))
                    text3= font2.render('chips left: ' +str(chip_two), 1, (237, 28, 36))
                    screen.blit(text3, (850, 210))
                    pygame.display.update()
                    pygame.time.delay(3000)
                gameWon = gameResult(move,player)
                draw_board(move)
                pygame.display.update()
                pygame.time.delay(3000)
                for i in range(9):
                    for j in range(1):
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size,300, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size,0, square_size, square_size))
                        pygame.draw.rect(screen, LIGHT_BLUE, (i*square_size+square_size+square_size, j+2*square_size, square_size, square_size))
                pygame.display.update()
                chip_two = chip_two - 1 + skipped_chip_two
                print("Player TWO has ", chip_two," chips left", "\n")
                skipped_turn_two += skipped_chip_two
                if gameWon == True:
                    player_two_win_count +=1
                    break
                else:
                    gameWon = False

        turn += 1
        turn = turn % 2
    print("Player one game length: " + str(countOne))
    print("Player two game length: " + str(countTwo))
    print("Player One win Count: " + str(player_one_win_count))
    print("Player Two win Count: " + str(player_two_win_count))
    print("Player one skipped turns length: " + str(skipped_turn_one))
    print("Player two skipped turns length: " + str(skipped_turn_two))
    return countOne, countTwo,skipped_turn_one,skipped_turn_two, player_one_win_count, player_two_win_count, draw_count
    pygame.quit()
    quit()
if __name__=='__main__':
    player_one_rule_ten_decision = input("Welcome player one, enter rule 10 decision type e.g lowest probability, highest probability, random: ").lower()
    player_one_rule_eleven_decision = input("Enter rule 11 decision type e.g lowest probability, highest probability, random: ").lower()
    player_two_rule_ten_decision = input("Welcome player two, enter rule 10 decision type e.g lowest probability, highest probability, random: ").lower()
    player_two_rule_eleven_decision = input("Enter rule 11 decision type e.g lowest probability, highest probability, random: ").lower()
    gamePlayTwoPlayerOneDimensionRule(player_one_rule_ten_decision, player_one_rule_eleven_decision,player_two_rule_ten_decision,player_two_rule_eleven_decision)
```
